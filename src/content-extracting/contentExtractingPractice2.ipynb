```python
import easyocr
from pdf2image import convert_from_path
from PIL import Image
from PIL import ImageEnhance
```

```python
def adjust_sharpness(input_image, output_image, factor):
    image = Image.open(input_image)
    enhance_object = ImageEnhance.Sharpness(image)
    out = enhance_object.enhance(factor)
    out.save(output_image)
```

```python
adjust_sharpness('./images/bachelor.degree.cnvh.png', './images/sharp-bachelor.degree.cnvh.png', 3)
```

```python
reader = easyocr.Reader(['vi', 'en'])
```

```python
result = reader.readtext('./images/sharp-bachelor.degree.cnvh.png', detail=True)
```
